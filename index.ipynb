## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
    SQL is a relational database that has the ability to store and retrieve data for analysis. It works well with data that is tabular in nature. NoSQL works well with data that cannot be stored adquately on a table. The one that works better depends on what type of data you are working with. For example, chat logs would work better on NoSQL while small amounts of tabular data work well on SQL.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
MongoDB is a type of NoSQL database that is good at storing document stores or data stored as a document. It can store data that is arbitarily long or short in key-value pairs. It is flexible because each document doesn't necessarily need to have the same amount of keys and work best when we don't know the shape of the document.







## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
Fault tolerant is a fail-safe where we have a way to store large amounts of data safely on multiple servers. For example, let's say we have an entire database stored on one server and that server goes down and we would have to wait for the server to go back up in order to continue. With RDDs, we can have mulitple servers that work with subsets of our data and back-ups of all the other subsets. This keeps our data safe and helps servers run faster.








## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
Map reduce is a technique where we run a query on different sets of servers and combine all the queries together. MapReduce is how a database like Hadoop works. Using multiple smaller servers to store different parts of a database and then bringing them all together.
